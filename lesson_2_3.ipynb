```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, num=51)
x
```

```python
y1 = x**2
y1
```

```python
y2 = 2*x + 0.5
y2
```

```python
y3 = -3*x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6))
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'red')
plt.title('График 1')
plt.xlim(-5, 5)
plt.subplot(2, 2, 2)
plt.plot(x, y2,'green')
plt.title('График 2')
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'yellow')
plt.title('График 3')
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('График 4')
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()

```
